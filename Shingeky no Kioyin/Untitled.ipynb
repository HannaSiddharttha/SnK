```python
website_url= rq.get('https://www.gamerfocus.co/anime/shingeki-no-kyojin-attack-on-titan-temporada-final-parte-1-resena/').text 
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url,'lxml')

data = [element.text for element in soup.find_all("p")]
data = str(data)
data
```

```python
from stop_words import get_stop_words

stop_words_es = get_stop_words('es')
```

```python
from wordcloud import WordCloud, STOPWORDS

print ('Wordcloud is installed and imported!')
```

```python
stop_words = ["eren", "mikasa", "zeke", "titan", "shingeky", "bestia", "gaby"] + stop_words_es
snk_wc = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stop_words
)

snk_wc.generate(data)
```

```python
plt.imshow(snk_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
snk_wc.generate(data)
fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18)

plt.imshow(snk_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
